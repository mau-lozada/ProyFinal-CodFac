Primero traeremos nuestras bases de datos. En este caso fueron 3 recuperadas de fuentes como INEGI y la ENOE.
Para esto activaremos nuestras librerias para poder leer archivos CSV, en este caso la librería de Pandas nos ayudará en esta parte.

De acuerdo a la INEGI "PEA" significa Población Economicamente Activa y "PD" es Población Desocupada.

```python
#Llamamos a Pandas para poder leer los archivos CSV, y le pondremos de nombre pd, como convencionalmente se usa en la comunidad.

import pandas as pd 
import numpy as np

desocupacion = pd.read_csv('desocupacion.csv')
desocupacion

```

Comenzamos con la visualización efectiva de los datos.

Ahora extraemos el data frame de Ocupación Informal. La Ocupación Informal son aquellos negocios remunerado o autoempleos que no están registrados, regulados o protegidos por marcos legales o normativos. Los trabajadores informales no cuentan con contratos de empleo, seguros, prestaciones laborales, protección social o representación de los trabajadores.

```python
informal = pd.read_csv('O_informal.csv')

informal.round(2)
```

Agregamos una columna, sacando el promedio en total por cada Periodo, de cada trimestre.

```python
informal['promedio'] = (informal['Hombres'] + informal['Mujeres']) / 2 #Creamos una nueva columna sacando el promedio.
informal.round(2) #redondeamos los decimales resultantes a sólo dos decimales.
```

Una vez teniendo toda esta información procedemos a visualizar los datos de una forma más dinamica con ayuda de Altair.

```python
import altair as alt #Importamos la librería de altair y le asignamos el nombre de "alt"
desocupacion2.head() #Llamamos los primeros datos del dataframe.
```

Sacamos la gráfica de la tasa del desempleo

```python
#slider = alt.binding_range(min=2015, max=2022, step=1) #Slider para poder desplazar el gráfico por año
#select_year = alt.selection_single(name='año', fields=['año'],
#                                     bind=slider, init={'año': 2015}) 
 
base = alt.Chart(desocupacion2).encode( # definimos 'base' para no repetir cada vez 'alt.Chart(data)'
    color = alt.Color("tasa", type = "nominal"),
    )#.add_selection(select_year).transform_filter(select_year)

left = base.mark_bar(width= 60).encode( #definimos la variable como "left", para usarla más tarde
    x='año',
    y="tasa",
    # The highlight will be set on the result of a conditional statement
    color=alt.condition(
        alt.datum.año == 2020,  # Si el año es 2020 lo retorna como True,
        alt.value('#fdc086'),   # Que returna la barra de color naranja.
        alt.value('#beaed4')    # y si no es True, retorna un valor purpura.
    )
).properties(width=600)

```

Sacamos la gráfica de los empleos informales

```python
base2 = alt.Chart(informal).encode() 
    
right = base2.mark_line().encode(   #Este comando es para un grafico de lineas y definimos la variable como "right", para usarla más tarde
    x=alt.X("año"),                 #Definimos que en el eje X será año
    y=alt.Y(
        alt.repeat("layer"), aggregate="mean", title="" #definimos que en el eje Y se calcule el promedio
    ),
    color=alt.datum(alt.repeat("layer")), #Esta parte se tomo de la documentación de Vega-Lite.
).properties(width=800).repeat(
    layer=["Hombres", "Mujeres"])
```

```python
final = alt.hconcat(left, right).resolve_scale(x = 'independent') #resolve_scale recibe valores como shared o independent para poder concatenar dos gráficos y las etiquetas del gráfico.
final
```

En este punto podemos observar que en el año 2020 a causa de la pandemia hubo un incremento en el desempleo a comparación que los 5 años antriores, pero paralelamente, para el año 2021 el empleo informal creció y se mantuvo en el siguiente año. Podemos afirmar que cuando varias personas quedaron desempleadas, decidieron poner un negocio por su cuenta y obtener ingresos independientemente.
